```python

"""
### BULGARIA CUENTA CON APROXIMADAMENTE CERCA DE LOS 6 MILLONES DE HABITANTES y lo que hize fue dividir a bulgaria en 3 zonas geograficas

###espero que le funcione profe :(
### EXplicacion
Visualizar cmo se distribuye la poblacin de Bulgaria en diferentes provincias agrupadas por zonas geogrficas,
 (norte, centro, sur).



###
 -Se definen **clases**:
   `Province`: representa una provincia con su nombre y poblacin.
   `Region`: agrupa varias provincias y calcula la poblacin total.
 -Se cargan los datos de poblacin por provincia .
  -Se agrupan las provincias en tres regiones: norte, centro y sur.
-Se genera un **grfico de barras apiladas (stacked bar chart)** usando **Matplotlib**.
-Se aaden etiquetas con el total de habitantes por regin.

### Resultado
El grfico final muestra 3 barras (una por regin), con secciones apiladas que representan
las provincias que pertenecen a cada zona.
"""

# jijijija 0.0
# aca comenzamos importando bien loquillo

import matplotlib.pyplot as plt
import pandas as pd

# jijijia 0.1
# Clases (


class Province:
    """Representa una provincia con su nombre y poblacin."""
    def __init__(self, name, population):
        self.name = name
        self.population = population

class Region:
    """Agrupa provincias y calcula la poblacin total."""
    def __init__(self, name, province_names, data_dict):
        self.name = name
        self.provinces = [Province(p, data_dict[p]) for p in province_names]

    def total_population(self):
        return sum(p.population for p in self.provinces)


# Datos de poblacin (aproximados humildemente)

province_data = {
    "Sofia": 1339000,
    "Plovdiv": 666000,
    "Varna": 474000,
    "Burgas": 380000,
    "Stara Zagora": 312000,
    "Pleven": 243000,
    "Veliko Tarnovo": 221000,
    "Ruse": 206000,
    "Blagoevgrad": 300000,
    "Pazardzhik": 238000,
    "Haskovo": 210000,
    "Shumen": 171000,
    "Sliven": 179000,
    "Dobrich": 156000,
    "Kardzhali": 150000,
}

# jijijijijja
# REGIONES PROFE MELQUI AQUI

regions_definition = {
    "Norte": ["Varna", "Ruse", "Pleven", "Veliko Tarnovo", "Dobrich", "Shumen"],
    "Centro": ["Sofia", "Plovdiv", "Stara Zagora", "Pazardzhik", "Sliven"],
    "Sur": ["Burgas", "Blagoevgrad", "Haskovo", "Kardzhali"],
}


# Crear objetos Region

regions = [Region(name, provs, province_data) for name, provs in regions_definition.items()]

# jijijija 2.0
# el grfico con Matplotlib

fig, ax = plt.subplots(figsize=(9,6))

# barras del grafico
for r in regions:
    bottom = 0
    for prov in r.provinces:
        ax.bar(r.name, prov.population, bottom=bottom, label=prov.name)
        bottom += prov.population

# Ttulo eje X e Y jijija
ax.set_title("Poblacion dividad en Norte, centro y SUR ", fontsize=14, fontweight='bold')
ax.set_ylabel("Poblacin (habitantes en Mill)")
ax.set_xlabel("Regin")

# cheque este datallito en las barras
for r in regions:
    ax.text(r.name, r.total_population() + 50000, f"{r.total_population():,}".replace(",", "."),
            ha='center', va='bottom', fontsize=10, fontweight='bold')

# Leyenda (solo provincias nicas)
handles, labels = ax.get_legend_handles_labels()
unique = dict(zip(labels, handles))
ax.legend(unique.values(), unique.keys(), title="Provincias", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()


#  Mostrar tabla resumen
df = pd.DataFrame({
    "Regin": [r.name for r in regions],
    "Poblacin total": [r.total_population() for r in regions]
})
print("=== Poblacin total por regin ===")
display(df)

#firma del jijija david caamao
```
